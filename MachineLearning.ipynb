```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data from your table
data = {
    'Adj Close': [8.768, 0.040, 149.399, 23.907, 571.554, 0.031, 149.430],
    'Close': [8.795, 0.361, 149.396, 23.905, 571.443, 0.034, 149.430],
    'High': [8.957, 0.357, 153.094, 24.350, 592.904, 0.036, 153.130],
    'Low': [8.618, 0.359, 147.787, 23.419, 548.459, 0.033, 147.820],
    'Open': [8.796, 0.040, 153.995, 23.923, 572.296, 0.035, 153.030],
    'Volume': [599110328.7, 451728000.0, 9211176000.0, 430723588.8, 1.855e+17, 19680000.0, 9230856000.0]
}

stats = pd.DataFrame(data, index=['Mean', 'Mode', 'Range', 'Std Dev', 'Variance', 'Min', 'Max'])

# Plot settings
plt.figure(figsize=(15, 12))
plt.suptitle('Descriptive Statistics Comparison', y=1.02, fontsize=16)

# Plot Mean and Mode
plt.subplot(2, 2, 1)
stats.loc[['Mean', 'Mode']].T.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Mean vs Mode')
plt.ylabel('Value')
plt.xticks(rotation=45)

# Plot Range, Min, Max
plt.subplot(2, 2, 2)
stats.loc[['Range', 'Min', 'Max']].T.plot(kind='bar', color=['lightgreen', 'lightblue', 'orange'])
plt.title('Range, Min, and Max')
plt.xticks(rotation=45)

# Plot Std Dev and Variance (log scale for variance)
plt.subplot(2, 2, 3)
stats.loc[['Std Dev']].T.plot(kind='bar', color='gold', legend=False)
plt.title('Standard Deviation')
plt.xticks(rotation=45)

ax2 = plt.subplot(2, 2, 4)
stats.loc[['Variance']].T.plot(kind='bar', color='orchid', ax=ax2, legend=False)
plt.yscale('log')  # Log scale due to huge variance values
plt.title('Variance (Log Scale)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Volume-specific plot (separate due to scale differences)
vol_stats = stats['Close'].drop('Variance')  # Exclude extreme variance
plt.figure(figsize=(10, 5))
vol_stats.plot(kind='bar', color='purple')
plt.title('Close Statistics (Excluding Variance)')
plt.ylabel('Count')
plt.yscale('log')  # Log scale for better visibility
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Calculate descriptive stats
stats = pd.DataFrame({
    'Mean': df.mean(),
    'Mode': df.mode().iloc[0],
    'Range': df.max() - df.min(),
    'Std Dev': df.std(),
    'Variance': df.var(),
    'Min': df.min(),
    'Max': df.max()
})
```

```python
plt.figure(figsize=(15, 10))

# Plot Mean
plt.subplot(2, 3, 1)
stats.loc['Mean'].plot(kind='bar', color='skyblue', title='Mean Values')

# Plot Mode
plt.subplot(2, 3, 2)
stats.loc['Mode'].plot(kind='bar', color='salmon', title='Mode Values')

# Plot Range
plt.subplot(2, 3, 3)
stats.loc['Range'].plot(kind='bar', color='lightgreen', title='Range (Max - Min)')

# Plot Standard Deviation
plt.subplot(2, 3, 4)
stats.loc['Std Dev'].plot(kind='bar', color='gold', title='Standard Deviation')

# Plot Variance
plt.subplot(2, 3, 5)
stats.loc['Variance'].plot(kind='bar', color='orchid', title='Variance')

# Plot Min & Max Together
plt.subplot(2, 3, 6)
df_stats_minmax = stats.loc[['Min', 'Max']].T
df_stats_minmax.plot(kind='bar', color=['lightblue', 'orange'], title='Min vs Max')

plt.tight_layout()
plt.suptitle('Descriptive Statistics Across Columns', y=1.02, fontsize=14)
plt.show()
```
